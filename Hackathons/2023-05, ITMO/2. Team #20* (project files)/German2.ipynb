```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.dpi': 500, 'figure.autolayout': True, 'figure.figsize': (9, 9)})
```

```python
df = pd.read_excel('./База данных.xlsx')
df.head()
```

```python
sns.set(rc={'figure.dpi': 500, 'figure.autolayout': True, 'figure.figsize': (3, 9)})
ax = sns.boxplot(data=df[df.fuel == 1].price, )
ax.set(title='Fuel service prices', xlabel='Fuel Price', ylabel='Price (RUB)')


plt.savefig('fuel service prices')
ax
```

```python
sns.set(rc={'figure.dpi': 500, 'figure.autolayout': True, 'figure.figsize': (9, 9)})
data = data=df[df.fuel == 1]
data.sort_values('price')
ax = sns.displot(data=data.price)
ax.set(title='Distribution of fuel prices', xlabel='Cost (RUB)', ylabel='Frequency')
ax.fig.set_dpi(500)
ax.fig.set_figwidth(5)
ax.fig.set_figheight(5)

plt.savefig('distribution of fuel prices')
ax
```
