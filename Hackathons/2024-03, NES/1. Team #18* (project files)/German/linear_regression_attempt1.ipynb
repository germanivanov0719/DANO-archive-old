```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('ecology_with_new_index.csv')
df.head()
```

```python
# idk
df['b5'][df['b5'] == 'Не знаю'] = df[df['b5'] != 'Не знаю']['b5'].astype(int).mean()
ind = df
ind['b7a'][1259] = 0
```

```python
X_train = pd.concat([pd.get_dummies(df.iloc[:, 1]).astype(int), 
                     pd.get_dummies(df.iloc[:, 2]).astype(int), 
                     pd.get_dummies(df.iloc[:, 3]).astype(int), 
                     df['b5'].astype(int), 
                     ind['b7a'].astype(int), 
                     ind['BMj5'].astype(int)], axis='columns')
y_train = ind.iloc[:, -1]
X_train_sm = sm.add_constant(X_train)
```

```python
lr = sm.OLS(y_train, X_train_sm).fit()
lr.summary()
```
