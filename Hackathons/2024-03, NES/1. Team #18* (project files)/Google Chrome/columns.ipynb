```python
cols = pd.read_excel("columns.xlsx")
cols = cols.dropna()
cols["id"] = cols['id'].astype(int)
cols
```

```python
alph = "абвгдеёжзийклмнопрстуфхцчшщъыьэюя"
lastOne = 1
for id in cols['id'].values:
    data = cols[cols['id'] == id]['strr'].values[0]
    first_word = data.split()[0]
    ok = True
    for let in alph:
        if let in first_word:
            ok = False
            break
    if not ok:
        # print(cols.loc[cols['id'] == lastOne, 'strr'])
        cols.loc[cols['id'] == lastOne, 'strr'] = cols.loc[cols['id'] == lastOne, 'strr'].values[0] + " " + data
        # print(f'changed to {cols[cols["id"] == lastOne]["strr"].values[0] + data}')
        cols = cols.drop(cols[cols['id'] == id].index)
        id = 1
    lastOne = id
```

```python
cols['name1'] = cols['strr'].apply(lambda x: x.split()[0] if type(x) == str else 0)
cols['discr'] = cols['strr'].apply(lambda x: " ".join(x.split()[1:]) if type(x) == str else 0)
cols
```

```python
for id in cols['id']:
    data[cols[cols['id'] == id]['name1'].values[0]] = [cols[cols['id'] == id]['discr'].values[0]]
```
