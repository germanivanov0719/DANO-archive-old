```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./sleep.xlsx')
df.head()
```

```python
fig, ax = plt.subplots()
sns.boxplot(data=df, x='occupation', y='sleepduration', ax=ax)
```

```python
df.groupby('qualityofsleep').count().personid.plot(kind='pie')
```

```python
sns.scatterplot(df, x='age', y='sleepduration')
```
